## Get Sankey figures from pre produced excel data files

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set();
import plotly.graph_objects as go
# Custom sankey functions for tour progress
import stpf

# Data produced from profile_tour.m MATLAB script
fn_et = "Sankey Data EUR Min 10.xlsx"
fig_et = stpf.produce_tour_progression_sankey(fn_et)

fn_pga = "Sankey Data USA Min 10.xlsx"
fig_pga = stpf.produce_tour_progression_sankey(fn_pga)
```

# Plot both figures


```python
fig_et.update_layout(title_text="European Tour 2015 - 2020")
fig_et.show()
fig_pga.update_layout(title_text="PGA TOUR 2015 - 2020")
fig_pga.show()
```
